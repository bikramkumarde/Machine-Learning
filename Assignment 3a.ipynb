```python
TU_10a = { 1 : {"transition" : "any",
               "reward" : -1,
               "next state" : "11aClass"
               }
        }

RU_10a = { 1 : {"transition" : "any",
               "reward" : 0,
               "next state" : "11aClass"
               }
        }

RD_10a = { 1 : {"transition" : "any",
               "reward" : 4,
               "next state" : "11aClass"
               }
        }

TD_10a = { 1 : {"transition" : "any",
               "reward" : 3,
               "next state" : "11aClass"
               }
         }
```

```python
states = {"RU_8p" : RU_8p,
          "TU_10p" : TU_10p,
          "RU_10p" : RU_10p,
          "RD_10p" : RD_10p,
          "intermediate1" : intermediate1,
          "intermediate2" : intermediate2,
          "RU_8a" : RU_8a,
          "RD_8a" : RD_8a,
          "TU_10a" : TU_10a,
          "RU_10a" : RU_10a,
          "RD_10a" : RD_10a,
          "TD_10a" : TD_10a
         }
```

```python
def func_rl(current_state, reward_count, output_str):
    if (current_state != '11aClass'):
        random_number = random.random()
        possibilities = len(states[current_state])
        chosen_step = math.ceil(possibilities*random_number)
        next_state = states[current_state][chosen_step]["next state"]
        if (next_state == 'intermediate1' or next_state == 'intermediate2'):
            step = states[current_state][chosen_step]["transition"]
            random_number = random.random()
            possibilities = len(states[next_state])
            chosen_step = math.ceil(possibilities*random_number)
            next_state1 = states[next_state][chosen_step]["next state"]
            current_state = next_state1
            current_reward = states[next_state][chosen_step]["reward"]
            output_str += (current_state + ' --> '+ step + ' --> '+ str(current_reward) + ' --> ')
            reward_count += current_reward
            
        else:
            step = states[current_state][chosen_step]["transition"]
            current_reward = states[current_state][chosen_step]["reward"]
            output_str += (current_state + ' --> '+ step + ' --> '+ str(current_reward) + ' --> ')
            reward_count += current_reward
            current_state = next_state
        
    return current_state, reward_count, output_str
        
```

```python
reward_sum = 0
for i in range(0,50):
    print('Step : {}'.format(i+1))
    count = 0
    str1 = ''
    current_state = 'RU_8p'
    
    while (current_state != '11aClass'):
        current_state, count, str1 = func_rl(current_state, count, str1)
    str1 += current_state
    print(str1 + ' ==> '+str(count) + '\n')
    reward_sum += count
    
print('Average Reward: {}' .format(reward_sum/50))
```
